В абсолютних значеннях рефандів і чарджбеків більше на контрольній групі, але треба враховувати, що вона становить 70% трафіку, тому при нормалізації стає зрозумілим, що пропорційно на тестовій групі кількість рефандів і чарджбеків більша = незадоволеність юзерів більша.

Будемо джойнити чарджбеки, рефанди і апсели на юзерів

```python
control_refs_cbs = df_control.merge(refs_and_cbs, how='left', left_on=["user_uuid", "Product Name"], right_on=["user_id", "product_name"]).merge(upsells_data, how='left', left_on='user_uuid', right_on='uuid')
```

```python
test_refs_cbs = df_test.merge(refs_and_cbs, how='left', left_on=["user_uuid", "Product Name"], right_on=["user_id", "product_name"]).merge(upsells_data, how='left', left_on='user_uuid', right_on='uuid')
```

```python
test_refs_cbs
```

```python
pd.isna(test_refs_cbs)[["chargeback", "refund"]].sum()
```

В фінальному підрахунку юзерів, які в подальшому зробили рефанд не враховуємо, а кількість юзерів з чарджбеками при сумуванні доходу множимо на -30.

```python
final_control_refs_cbs = control_refs_cbs[control_refs_cbs["refund"] == 0]
final_test_refs_cbs = test_refs_cbs[test_refs_cbs["refund"] == 0]
```

```python
final_control_refs_cbs
```

```python
revenue_by_lt_control = final_control_refs_cbs.groupby(["Date Started", "rebill_lt"], as_index=False).agg(
    unique_users=("user_uuid", "nunique"),
    chargeback_users=("chargeback", "sum")
)
revenue_by_lt_test = final_test_refs_cbs.groupby(["Date Started", "rebill_lt"], as_index=False).agg(
    unique_users=("user_uuid", "nunique"),
    chargeback_users=("chargeback", "sum")
)
```
